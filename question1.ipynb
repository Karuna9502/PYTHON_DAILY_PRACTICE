```python
import numpy as np
```

```python
1. #CREATE A NULL ARRAY OF ANY LENGTH WITH THE MIDDLE VALUE AS A NATURAL NUMBER
n = int(input("enter array lenght : - "))
num = int(input("enter any random number"))
if n%2 == 0:
  mid = n/2
else:
  mid = n//2
arr = []
for i in range(n):
  if i == mid:
    arr.append(num)
  else:
    arr.append(0)
arr = np.array(arr)
print(arr)
```
