**Find Maximum of two numbers in Python**

```python
# USING MAX() FUNCTION IT IS VERY EASY TO USE AND QUICK METHOD
a = 10
b = 13
print(max(a,b))
```

```python
#AGAIN USING PYTHON SORT() FUNCTION
#he two numbers are stored in a list, which is then sorted to find the maximum. Although this method works correctly,
# it introduces unnecessary computational overhead,
# making it inefficient for such a simple task.

a = 87
b = 100
num = [a, b] #storing this in list increasing steps
num.sort()
print(num[-1])
# Using num[-1] accesses the maximum value. Since 100 is greater than 87, the list becomes [87, 100] and num[-1] returns 100.
```

```python
# USING TERNARY OPERATOR
a = 28
b = -7
print(a if a > b else b)
```

```python
#USING IF ELSE STATEMENT
a = 67
b = 98
if a > b:
  print(a)
else:
  print(b)
```
