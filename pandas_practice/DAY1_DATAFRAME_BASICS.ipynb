**Creating a Pandas DataFrame**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
#Creating an Empty DataFrame
df = pd.DataFrame
print(df)
```

```python
#Creating DataFrame from dict of Numpy Array
data = {'A': np.array([1,2,3]),
        'B': np.array([4,5,6]),
        "C": np.array([7,8,9])}

df = pd.DataFrame(data)
print(df)

#Each key in the dictionary represents a column name and the corresponding
# NumPy array provides the values for that column.
```

```python
#Creating a DataFrame from a List of Dictionaries

import pandas as pd

data = [
    {'name':"karuna", "degree":"MCA", "age":32},
    {'name': 'anisha', "degree":"MBA", "age":24},
    {'name':'ankita', "degree":"no-degree", "age":16}
]

df = pd.DataFrame(data)
print(df)
```
