**CREATE AN EMPTY AND FULL NUMPY ARRAY**

```python
import numpy as np
```

**AS WE KNOW NUMPY HAVE IN-BULIT FUNCTION np.empty() and np.full()**




**I USE THESE FUNCTION TO SOLVE THIS PROBLEM**

```python
empty_array = np.empty((3,4))
print(empty_array)
```

```python
full_array = np.full((3,4), 3)
print(full_array)
```

```python
empty1_array = np.empty(5)
print(empty1_array)
```

```python
full1_array = np.full((3),5)
print(full1_array)
```
